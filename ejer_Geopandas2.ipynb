```python

import geopandas as gpd
import rasterio
import contextily as ctx
```

```python
muertes = gpd.read_file('cholera/Cholera_Deaths.shp')
```

```python
muertes
```

```python
bombas_de_agua = gpd.read_file('cholera/Pumps.shp')
bombas_de_agua
```

```python
#plotear el GEO DF
ax = muertes.plot(column='Count', alpha=0.5, edgecolor='k', legend=True)
```

```python
# agregarle bombas al mapa
ax = muertes.plot(column='Count', figsize=(10, 10), alpha=0.5, edgecolor='k', legend=True)
bombas_de_agua.plot(ax=ax, marker='x', color='red', markersize=50)
```

```python
#generar una sola fuente para 

ax = muertes.plot(column='Count', figsize=(15, 15), alpha=0.5, edgecolor='k', legend=True)
bombas_de_agua.plot(ax=ax, marker='x', color='red', markersize=50)

ctx.add_basemap(
    ax,
    crs=muertes.crs.to_string(),
)
```

```python
#en mapa de la Ã©poca

ax = muertes.plot(column='Count', figsize=(15, 15), alpha=0.5, edgecolor='k', legend=True)
bombas_de_agua.plot(ax=ax, marker='x', color='red', markersize=50);

ctx.add_basemap(ax,
    crs=muertes.crs.to_string(),
    source="cholera/SnowMap.tif"
)
```
